```python

url = 'https://api.binance.com/api/v3/klines'

params = {
    'symbol': 'BTCUSDT',
    'interval': '8h',
    'limit': 1000
}

btc_prices = []

response = requests.get(url, params=params)
response.raise_for_status()  # Raise an exception for HTTP errors
data = response.json()
for entry in data:
    price_time = datetime.fromtimestamp(entry[0] / 1000).strftime('%Y-%m-%d-%H')
    btc_info = {
        'DateTime': price_time,
        'BTC Price': float(entry[4])  # Closing price
    }
    btc_prices.append(btc_info)
btc_prices
```
