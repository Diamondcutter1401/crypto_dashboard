```python
import csv
import pandas as pd
import matplotlib.pyplot as plt

btc_df = pd.read_csv('btc_price_binance.csv')
open_interest_df = pd.read_csv('open_interest_binance.csv')
btc_df.rename(columns={'Close':'BTC Price','Open Time':'Date'}, inplace=True)
```

```python
btc_df
```

```python
open_interest_df
```

```python

merged_df = pd.merge(open_interest_df,btc_df,left_on='Date',right_on='Date',how='left')
merged_df
```

```python

merged_df.to_csv('merged_open_interest_to_btc.csv', index=False)

# Plot BTC Price on the primary y-axis
fig, ax1 = plt.subplots(figsize=(14, 7))
ax1.set_xlabel('DateTime')
ax1.set_ylabel('BTC Price (USD)', color='tab:blue')
ax1.plot(merged_df['Date'], merged_df['BTC Price'], color='tab:blue', label='BTC Price')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a secondary y-axis for the Open Interest
ax2 = ax1.twinx()
ax2.set_ylabel('Open Interest', color='tab:orange')
ax2.plot(merged_df['Date'], merged_df['Open Interest'], color='tab:orange', label='Open Interest')
ax2.tick_params(axis='y', labelcolor='tab:orange')

# Title and legends
plt.title('Open Interest and BTC Price Over Time')
fig.tight_layout()  # Adjust layout to make room for the labels
plt.show()
```
